```python
import numpy as np
import matplotlib.pyplot as plt

def run_hubble_phase_transition():
    print("LCT Simulation H: Vacuum Phase Transition (Early Dark Energy)")
    
    # 1. Cosmic Timeline (Redshift z converted to Scale Factor a)
    # a = 1 / (1 + z). Early Universe (a -> 0), Today (a = 1)
    a = np.linspace(0.001, 1.0, 1000)
    
    # 2. Standard LambdaCDM Prediction (Constant Dark Energy Density)
    # This leads to the "CMB" measurement of ~67 km/s/Mpc
    H_cmb = 67.4
    h_standard = np.ones_like(a) * H_cmb
    
    # 3. LCT Phase Transition Model
    # Instead of linear drift, we model a "Lattice Settling Event" (Phase Transition).
    # This releases a burst of vacuum energy (Early Dark Energy), bumping H0.
    # We use a Sigmoid function to model the phase transition.
    
    H_late = 73.0  # Supernova measurement
    transition_scale = 0.1 # When did it happen? (a=0.1 corresponds to z=9, Reionization)
    width = 0.05   # How fast was the freeze?
    
    # Sigmoid Logic: H(a) transitions from H_cmb to H_late
    h_lct = H_cmb + (H_late - H_cmb) / (1 + np.exp(-(a - transition_scale)/width))
    
    # 4. Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(a, h_standard, 'b--', linewidth=2, label='Standard LambdaCDM (Constant Vacuum)')
    plt.plot(a, h_lct, 'r-', linewidth=3, label='LCT Vacuum Phase Transition')
    
    plt.fill_between(a, h_standard, h_lct, color='purple', alpha=0.1, label='Hubble Tension Gap')
    
    plt.text(0.5, 69, "Vacuum Phase Transition\n(Lattice Settling)", color='red', ha='center')
    plt.arrow(0.2, 70, 0.1, 0, head_width=0.5, head_length=0.05, fc='red', ec='red')
    
    plt.title("Anomaly II: Solving H0 via Vacuum Phase Transition")
    plt.xlabel("Scale Factor (a) [Time ->]")
    plt.ylabel("Effective H0 (km/s/Mpc)")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.savefig('hubble_phase_transition.png', dpi=300)
    plt.show()

if __name__ == "__main__":
    run_hubble_phase_transition()
```
